```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from synthetic_task.plot_results import load_results, plot_total_loss_vs_method_by_ydim, plot_total_time_vs_method_by_ydim, plot_total_time_vs_ydim_for_methods
```

```python
sns.set_theme(style="whitegrid", context="talk")
palette = sns.color_palette()
```

```python
# PLOT GENERAL TASK -- SOC

batch_size = 8
SOC_DIR = f"../synthetic_results_general_{batch_size}"
METHODS = [
    "cvxpylayer",
    "lpgd",
    "bpqp",
    "ffocp_eq",
]
METHODS_LEGEND = {
    "cvxpylayer": "CvxpyLayer",
    "lpgd": "LPGD",
    "bpqp": "BPQP",
    "ffocp_eq": "FFOCP",
}

METHODS_STEPS = [method+"_steps" for method in METHODS]

method_order = [METHODS_LEGEND[m] for m in METHODS]

markers = ["o", "s", "D", "^", "v", "x"]
markers_dict = {method: markers[i] for i, method in enumerate(method_order)}

LINEWIDTH = 1.5
```

```python
df = load_results(base_dir=SOC_DIR, methods=METHODS, methods_legend=METHODS_LEGEND)
df = df.rename(columns=lambda c: c.strip() if isinstance(c, str) else c)
df["method"] = pd.Categorical(df["method"], categories=method_order, ordered=True)
```

```python
plot_total_time_vs_method_by_ydim(df, SOC_DIR, ('forward_time','backward_time'), tag="syn")
```

```python
# plot all ydim
selected_methods = ["FFOCP", "BPQP"]
plot_total_time_vs_ydim_for_methods(
    df,
    methods_subset=selected_methods,
    time_names=('forward_time', 'backward_time'),
    plot_path=SOC_DIR,
    plot_name_tag="syn_steps_solve_ffocp_bpqp"
)
```

```python
df = load_results(base_dir=SOC_DIR, methods=METHODS_STEPS)
df = df.rename(columns=lambda c: c.strip() if isinstance(c, str) else c)
df["method"] = pd.Categorical(df["method"], categories=method_order, ordered=True)
```
